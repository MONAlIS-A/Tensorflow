# Mathematical Operations


```python
X = tf.constant([1,2,3])
Y = tf.constant([9,8, 7])
```

```python
Z = tf.add(X, Y)
Z
```

```python
Z = X+Y
Z
```

```python
Z = tf.subtract(X, Y)
Z
```

```python
Z = tf.divide(X, Y)
Z
```

```python
Z = tf.multiply(X, Y)
Z
```

```python
Z = tf.tensordot(X, Y, axes=1)
Z
```

```python
Z = tf.reduce_sum(X*Y, axis=0)
Z
```

```python
Z = tf.reduce_sum(X*Y, axis=0)
Z
```

```python
Z = X**5
Z
```

```python
X = tf.random.normal((2,3))
Y = tf.random.normal((3,4))

Z = tf.matmul(X,Y)
Z
```

# Indexing

```python
X = tf.constant([0,1,1,2,3,1,2,3])
print(X[:])
```

```python
print(X[1:])
```

```python
print(X[1:3])
```

```python
print(X[::2])
```

```python
print(X[::-1])
```

```python
indices = tf.constant([0, 3])
X_ind = tf.gather(X, indices)
print(X_ind)
```

```python
X = tf.constant([[1,2],
                 [3,4],
                 [5,6]])
print(X[0,:])
print(X[0:2,:])
```

# Reshaping

```python
X = tf.range(9)
print(X)
```

```python
X = tf.reshape(X, (3,3))
print(X)
```

```python
X = tf.transpose(X, perm=[1,0])
print(X)
```
